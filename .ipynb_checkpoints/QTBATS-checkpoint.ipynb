```python
import pennylane as qml
from pennylane import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
```

```python
df = pd.read_csv('MTA_Subway_Ridership_Data_Clean.csv')
df["Date"] = pd.to_datetime(df["Date"], format="%m/%d/%Y")
df.set_index("Date", inplace=True)
df.rename(columns={
    "Subways: Total Estimated Ridership" : "Total Ridership",
    "Subways: % of Comparable Pre-Pandemic Day" : "Pre-Pandemic Ridership"
}, inplace=True)
```

```python
df['Total Ridership Standardized'] = (df['Total Ridership'] - df['Total Ridership'].mean()) / df['Total Ridership'].std()
```

```python
df.head()
```

```python
original_y = df['Total Ridership Standardized'].values.astype(float)
```

```python
N = 1 << (len(original_y) - 1).bit_length()
norm = np.linalg.norm(original_y)
y = np.pad(original_y, (0, N - len(original_y)), mode='constant')
y = y / norm
n_qubits = int(np.ceil(np.log2(N)))
```

```python
print(f"Number of qubits: {n_qubits}")
```

```python
dev = qml.device("default.qubit", wires=n_qubits, shots=None)
```

Implement the **Quantum Fourier Transform** circuit

```python
def QFT(wires):
    N = len(wires)
    for i in range(N):
        qml.Hadamard(wires=i)
        for j in range(i+1, N):
            qml.ControlledPhaseShift(np.pi/2**(j-i), wires=[j, i])
    for i in range(N//2):
        qml.SWAP(wires=[i, N-1-i])
```

```python
@qml.qnode(dev)
def getAmplitude():
    qml.StatePrep(y, wires=range(n_qubits), normalize=False)
    QFT(range(n_qubits))
    state = qml.state()
    return state
```

```python
def getCoeff(K):
    alpha = getAmplitude().tolist()[K]
    X = alpha * norm * np.sqrt(N)
    a = 2 / N * np.real(X)
    b = -2 / N * np.imag(X)
    return a, b
```

```python
def periodToBin(period, N):
    return round(N / period)
```

```python
weekly_bin, yearly_bin = periodToBin(7, N), periodToBin(365, N)
print(f"Weekly bin: {weekly_bin}, Yearly bin: {yearly_bin}")
```

```python
a_weekly, b_weekly = getCoeff(weekly_bin)
a_yearly, b_yearly = getCoeff(yearly_bin)
print(f"Weekly coefficients: a={a_weekly}, b={b_weekly}")
print(f"Yearly coefficients: a={a_yearly}, b={b_yearly}")
```

```python
def reconstructSeason(a, b, k, N, L):
    t = np.arange(L)  # reconstruct only over original series length
    S = a * np.cos(2 * np.pi * k * t / N) + b * np.sin(2 * np.pi * k * t / N)
    return S
```

```python
weekly_curve = reconstructSeason(a_weekly, b_weekly, weekly_bin, N, len(original_y))       # weekly period = 7
yearly_curve = reconstructSeason(a_yearly, b_yearly, yearly_bin, N, len(original_y))     # yearly period = 365
seasonal_curve = weekly_curve + yearly_curve
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df.index, original_y, label='Original', color='blue')
plt.plot(df.index, seasonal_curve, label='Seasonal', color='red')
plt.legend()
plt.show()
```

Now that we have obtained the `seasonal_curve` using **QFFT**, we can implement a **Damped Trend** for our pipeline.
We need a damped trend because our time period of interest is $2020$-$2025$, which includes the COVID pandemic and the recovery period.

The damped trend is usually formatted as $l_t = l_{t-1} + \phi b_{t-1} + \epsilon_t$ and $b_t = \phi b_t + \beta\epsilon_t$ -- the full ETS-style damped trend.

```python
def dampedTrend(residual, phi: float = 0.9, alpha: float = 0.1, beta: float = 0.01):
    N = len(residual)
    trend_curve = np.zeros_like(residual, dtype=float)
    # Initialize level and slope
    level = residual[0]
    slope = residual[1] - residual[0]
    trend_curve[0] = level

    for t in range(1, N):
        forecast = level + phi * slope
        error = residual[t] - forecast
        # Update states
        level = forecast + alpha * error
        slope = phi * slope + beta * error
        trend_curve[t] = level

    return trend_curve
```

```python
residual = y[:len(df)] - seasonal_curve
trend_curve = dampedTrend(residual)
arma_input = residual - trend_curve
```

We can add the **ARMA** term now

```python
def fitARMA(residual, p: int=2, q: int=2):
    model = ARIMA(residual, order=(p, 0, q))
    fit = model.fit()
    fitted = fit.fittedvalues
    return fitted, fit
```

```python
arma_fitted, arma_model = fitARMA(arma_input)
```

```python
y_hat = trend_curve + seasonal_curve + arma_fitted
plt.figure(figsize=(10, 5))
plt.plot(df.index, original_y, label='Original', color='blue')
plt.plot(df.index, y_hat, label='Fitted', color='red')
plt.legend()
plt.show()
```

```python
# Create a figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Plot ACF on the left subplot
plot_acf(arma_input, lags=50, ax=ax1)
ax1.set_title('Autocorrelation Function (ACF)')

# Plot PACF on the right subplot
plot_pacf(arma_input, lags=50, ax=ax2)
ax2.set_title('Partial Autocorrelation Function (PACF)')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
from itertools import product
import statsmodels.api as sm
```

```python
def auto_arma(y, p_max=6, q_max=6, criterion="aic"):
    """
    Grid search ARMA(p,q) using statsmodels, selecting best by AIC/BIC.
    """
    import warnings
    warnings.filterwarnings("ignore")
    best_ic = np.inf
    best_order = None
    best_model = None

    for p, q in product(range(p_max+1), range(q_max+1)):
        if p == 0 and q == 0:  # skip ARMA(0,0)
            continue
        
        model = sm.tsa.ARIMA(y, order=(p,0,q)).fit(method="statespace")
        ic = getattr(model, criterion)
        if ic < best_ic:
            best_ic = ic
            best_order = (p, q)
            best_model = model

    return best_model, best_order, best_ic
```

```python
model, order, score = auto_arma(arma_input)
```

```python
print("Selected ARMA order:", order)
print("Model score:", score)
print(model.summary())
```

Now we can get the fitted values

```python
fitted_values = model.fittedvalues
```

```python
arma_residuals = model.resid
```

```python
# Create a figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Plot ACF on the left subplot
plot_acf(arma_residuals, lags=50, ax=ax1)
ax1.set_title('Autocorrelation Function (ACF)')

# Plot PACF on the right subplot
plot_pacf(arma_residuals, lags=50, ax=ax2)
ax2.set_title('Partial Autocorrelation Function (PACF)')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

Here we can see that the **ACF** and **PACF** plots are mostly white noise, except for the giant spike at $\text{lag} = 0$, which is to be expected as the autocorrelation function of a series with itself is $1$.
